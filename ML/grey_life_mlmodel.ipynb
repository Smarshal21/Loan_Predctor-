```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/loan_approval_dataset - loan_approval_dataset.csv')
df
```

```python
df.info()
```

```python
df['education'].unique()
df.drop(columns=['loan_id'])
```

```python
#assgning the graduate as 1 and not graduate as 0
df['education']=df['education'].map({"Graduate":1,"Not Graduate":0})
```

```python
df['srit_score'].unique()
```

```python
#yes 1 and no 0
df['srit_score']=df['srit_score'].map({'Yes':1,'No':0})
```

```python
df.isnull().sum()
df['loan_status'].unique()
```

```python
#assigining approved 1 and rejected 0
df['loan_status']=df['loan_status'].map({'Approved':1, 'Rejected':0})
df.info()
df=df.drop(columns=['loan_id'])
```

```python
df.info()
x=df.iloc[:,0:11].values
y=df['loan_status']
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df[['no_of_dependents', 'income_annum', 'loan_amount', 'loan_term', 'cibil_score', 'loan_status']])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.75)
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit_transform(x_train)
sc.transform(x_test)
import seaborn as sns
import matplotlib.pyplot as plt
corr_matrix = df[['no_of_dependents', 'income_annum', 'loan_amount', 'loan_term', 'cibil_score', 'loan_status']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
import seaborn as sns
import matplotlib.pyplot as plt
sns.violinplot(x='loan_status', y='income_annum', data=df)
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
dtc.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
y_pred=dtc.predict(x_test)
cm=confusion_matrix(y_test,y_pred)
```
